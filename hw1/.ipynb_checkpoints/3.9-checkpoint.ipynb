#### Want to minimize $g(w) = w_0^2 + w_1^2 + 2sin(1.5(w_0 + w_1)) + 2$

```python
%matplotlib notebook
from matplotlib import rcParams
rcParams['figure.autolayout'] = True
import autograd.numpy as np   
import matplotlib.pyplot as plt
from autograd import grad
from collections import OrderedDict

init_point = np.array([[3.0,3.0]])
iterations = 10
alpha_1 = 1/100
alpha_2 = 1/10
alpha_3 = 1.0
# weight_history = init_point
g = lambda w: w[:,0]**2 + w[:,1]**2 + 2*np.sin(1.5*(w[:,0] + w[:,1])) + 2


def grad_descent(func, alpha, max_iterations, w):
    gfunc = grad(func)
    
    weight_history = np.array(w)
    cost_history = np.array(func(w))
    
    for k in range(max_iterations):
        grad_eval = gfunc(w)
        w = w - alpha*grad_eval
        
        weight_history = np.vstack([weight_history,w])
        cost_history = np.vstack([cost_history,func(w)])
    return weight_history, cost_history

a1_wh, a1_ch = grad_descent(g, alpha_1, iterations, init_point)
a2_wh, a2_ch = grad_descent(g, alpha_2, iterations, init_point)
a3_wh, a3_ch = grad_descent(g, alpha_3, iterations, init_point)
```

```python
init_point = np.array([[3.0,3.0]])
plt.plot(a1_ch, 'k-', label = 'alpha = .01')
plt.plot(a2_ch, 'm-', label = 'alpha = .1')
plt.plot(a3_ch, 'b-', label = 'alpha = 1')
plt.title("Cost Function History Plot")
plt.legend(loc="lower right")
plt.xlabel("step k")
plt.ylabel("g(w^k)")
```
