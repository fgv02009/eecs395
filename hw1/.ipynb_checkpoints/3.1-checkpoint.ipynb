a) 
$$ g(w) = \frac{1}{2}w^TBw + c^Tw, B = 
\begin{bmatrix}
  2 & 1 \\ 
  1 & 3
  \end{bmatrix}, c = 
  \begin{bmatrix}
  1 \\
  1
  \end{bmatrix}$$

$$\frac{dg}{dw} = w^TB + c^T$$

$$w^TB + c^T = 0$$

$$w^TB = -c^T$$

$$w^T = -c^TB^{-1}$$

$$\begin{bmatrix}
  2 & 1 \\ 
  1 & 3
  \end{bmatrix}^{-1} = \begin{bmatrix}
  \frac{3}{5} & \frac{-1}{5} \\ 
  \frac{-1}{5} & \frac{2}{5}
  \end{bmatrix}$$
  
$$w^T = \begin{bmatrix}
  -1 & -1
  \end{bmatrix}\begin{bmatrix}
  \frac{3}{5} & \frac{-1}{5} \\ 
  \frac{-1}{5} & \frac{2}{5}
  \end{bmatrix}$$
  
$$w = \begin{bmatrix}
  \frac{-2}{5} \\
  \frac{-1}{5}
  \end{bmatrix}$$

```python
w_stationary = np.array([-2/5, -1/5])
B = np.array([[2, 1], [1, 3]])
c = np.array([[1], [1]])
g = lambda w: np.dot(np.dot(.5*w.T,B),w) + np.dot(c.T,w)

w0 = np.linspace(-3,3,50)
w1 = np.linspace(-3,3,50)

# W0, W1 = np.meshgrid(w0, w1)

Z_input = np.array([w0, w1])
Z = g(Z_input)


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(w0, w1, Z, 50, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.scatter(w_stationary[0], w_stationary[1], g(w_stationary), 'g.', label = "stationary point")
plt.show()

#couldn't figure out how to plot point on top of 3d figure -- but stationary point is solved for above

# plt.figure()
# ax = plt.axes(projection='3d')
# ax.contour3D(w_stationary[0], w_stationary[1], g(w_stationary), 50, cmap=cm.coolwarm)
# ax.set_xlabel('x')
# ax.set_ylabel('y')
# ax.set_zlabel('z')
# plt.show()
# print(w_stationary[0])
```

b) 
$$g(w) = w\log(w) + (1-w)\log(1-w) = w\frac{\ln(w)}{\ln(10)} + (1-w)\frac{\ln(1-w)}{\ln(10)} = \frac{1}{\ln(10)}[w\ln(w) + (1-w)\ln(1-w)]$$

$$\frac{dg}{dw} = \frac{1}{\ln(10)}[1+\ln(w) + (-1)(\ln(1-w)) + (1-w)\frac{1}{1-w}(-1)]$$

$$\frac{dg}{dw} = \ln(w) - \ln(1-w)$$

$$\frac{dg}{dw} = \log(w) - \log(1-w)$$

$$\log(w) - \log(1-w) = 0$$

$$\log(w) = \log(1-w)$$

$$\log(w) = \frac{1}{2}$$

```python
plt.figure()
w_stationary = .5
g = lambda w: w*np.log(w) + (1-w)*np.log(1-w)

x = np.linspace(0,1,1000)
output = []
for input in x:
    output.append(round(g(input),2))
    
plt.plot(x,output, 'b-', label = "g")
plt.plot(w_stationary, g(w_stationary), 'r.', label = "stationary point")
# plt.plot(x,output_non_grad, 'r-', label = "g")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.annotate('stationary point is a minima', xy=(w_stationary, g(w_stationary)), xytext=(.8, -.7),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.show()
```

c)
$$g(w) = \log(1+e^w)$$

$$g(w) = \log(1+e^w) = \frac{\ln(1+e^w)}{\ln(10)}$$

$$\frac{\ln(1+e^w)}{\ln(10)} = \frac{1}{\ln(10)}(\ln(1+e^w))$$

$$\frac{dg}{dw}\frac{1}{\ln(10)}(\ln(1+e^w)) = \frac{1}{1+e^w}e^w =  \frac{e^w}{1+e^w}$$

$$\frac{dg}{dw} = \frac{e^w}{e^w + 1}$$

$$\frac{e^w}{e^w + 1} = 0$$

$$e^w = 0$$

$$\lim_{w\to\infty} = 0$$

```python
plt.figure()
w_stationary = -10
g = lambda w: np.log(1+np.exp(w))

x = np.linspace(-10,10,10000)
output = []
for input in x:
    output.append(round(g(input),2))
    
plt.plot(x,output, 'b-', label = "g")
plt.plot(w_stationary, g(w_stationary), 'r.', label = "stationary point at -infinity")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.annotate('stationary point as w approaches -infinity is a saddle point', xy=(w_stationary, g(w_stationary)), xytext=(-7.5, 2),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.show()
```

d) 
$$g(w) = w\tanh(w)$$

$$\frac{dg}{dw} = \tanh(w) + w(1-\tanh^2(w))$$

$$\frac{dg}{dw} = \tanh(w) + w - w\tanh^2(w)$$

$$\frac{dg}{dw} = \tanh(w) + w - w(1-\sech^2(w))$$

$$\frac{dg}{dw} = \tanh(w) + w - w + w\sech^2(w)$$

$$\frac{dg}{dw} = \tanh(w) + w\sech^2(w)$$

$$\tanh(w) + w\sech^2(w) = 0$$

$$\tanh(w) = -w\sech^2(w)$$

$$w = 0$$

```python
plt.figure()
w_stationary = 0
g = lambda w: w*np.tanh(w)

x = np.linspace(-5,5,1000)
output = []
for input in x:
    output.append(round(g(input),2))
    
plt.plot(x,output, 'b-', label = "g")
plt.plot(w_stationary, g(w_stationary), 'r.', label = "stationary point")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.annotate('saddle point is a minima', xy=(w_stationary, g(w_stationary)), xytext=(2, 0),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.show()
```
